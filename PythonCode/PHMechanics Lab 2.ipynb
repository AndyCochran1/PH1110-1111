```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy 

#This is a great website for getting started with plotting with pandas and matplot lib
#https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html
```

```python
lin_data = pd.read_csv("walking_linear.csv")
#print(lin_data) #Again, useful diagnostic tool to see if your data looks as expected. 
lin_data.dtypes #Excelent, I have all my variables (x,y,t). I had to delete the "mass_A" off the top of my file.  
#That was automatically copied over
```

```python
lin_data.plot('t') #This command plots x and y against t. Looks very similar to my Tracker plots. X is quite linear and y is 
#mostly constant. 
```

```python
line = numpy.polyfit(lin_data.t, lin_data.x, 1) #notice that it's t (time) then x (position). 1 refers to linear here. 
p = numpy.poly1d(line) #Do this to convert it into a usable form for plotting
```

### Plotting the position versus a linear trend

```python
fig=plt.figure() #initiate the figure
ax=fig.add_subplot(111) #So we can add multiple lines

ax.plot(lin_data.t,lin_data.x, c='b',marker="^",ls='--',label='x',fillstyle='none') #Data is horizontial axis, then vertical axis
#ax.plot(lin_data.t,lin_data.y,c='g',marker=(8,2,0),ls='--',label='y') #Saving time versus vertical position for later
ax.plot(lin_data.t, p(lin_data.t), c = 'r', ls = '--', label= p) #plotting the linear formula

#plt.text(2, 0.65, 'y = {} x + {}'.format(m, b)

plt.legend(loc=2) #Legends are very useful
plt.draw()
plt.title("Linear trend versus horizontial position")
ax.set_xlabel('time (s)')
ax.set_ylabel('position (m)')
print(p) #So you can see the form of the polynomial that was fitted. In this case it is y = mx + b

#plt.text? #another way to add annotations to a plot
```
